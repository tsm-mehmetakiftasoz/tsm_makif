```
# Gerekli kütüphaneler
import pandas as pd
from google.colab import files
import io

# Kullanıcıdan Excel dosyalarını seçmesini sağlama
uploaded_files = files.upload()  # Buradan birden fazla dosya seçebilirsin

# Boş bir DataFrame oluştur
combined_df = pd.DataFrame()

# Her dosyayı oku ve birleştir
for filename in uploaded_files.keys():
    df = pd.read_excel(io.BytesIO(uploaded_files[filename]))
    combined_df = pd.concat([combined_df, df], ignore_index=True)

# Birleştirilen DataFrame'i Excel olarak kaydet
output_filename = "birlesmis_excel.xlsx"
combined_df.to_excel(output_filename, index=False)

# Dosyayı Colab üzerinden indirilebilir hale getir
files.download(output_filename)
```
