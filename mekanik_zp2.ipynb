```
def temizle_ve_cevir(x):
    try:
        return str(int(float(x)))
    except:
        return str(x)
```

```
import pandas as pd

# Excel dosyasını yükle (örnek: "veri.xlsx")
df = pd.read_excel("/content/mekanik_zp.xlsx")
```

```
# 2️⃣ String olarak kalması gereken kolonlar
df["zp-zppoz"] = df["zp-zppoz"].astype(str).str.strip()
df["Партия"] = df["Партия"].astype(str).str.strip()

# 3️⃣ Кол-во в ЕИ ввода sütununu float'a çevir
df["Кол-во в ЕИ ввода"] = (
    df["Кол-во в ЕИ ввода"]
    .str.replace(".", "", regex=False)    # binlik ayırıcı noktaları kaldır
    .str.replace(",", ".", regex=False)   # ondalık virgülü noktaya çevir
    .str.replace(" ", "", regex=False)    # boşlukları kaldır
    .astype(float)                        # sayıya çevir
)

# 4️⃣ Gruplama için kolonlar
group_cols = ["zp-zppoz", "Партия"]

# 5️⃣ Her grup için toplam miktarı hesapla
sum_per_group = df.groupby(group_cols)["Кол-во в ЕИ ввода"].sum().reset_index()

# 6️⃣ Toplamı 0 olan grupları bul
zero_groups = sum_per_group[sum_per_group["Кол-во в ЕИ ввода"] == 0]

# 7️⃣ Orijinal DF'den bu grupları çıkar
mask = df.set_index(group_cols).index.isin(zero_groups.set_index(group_cols).index)
df_temiz = df[~mask].copy()

# 8️⃣ Sonucu kaydet
df_temiz.to_excel("/content/temizlenmis_veri.xlsx", index=False)

df_temiz
```
