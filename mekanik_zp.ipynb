```
import pandas as pd
import numpy as np
from google.colab import files
```

```
def temizle_ve_cevir(x):
    try:
        return str(int(float(x)))
    except:
        return str(x)
```

```
df_me2n_2022 = pd.read_csv('/content/ME2N_2022.csv', sep=None, engine='python')
df_me2n_2023 = pd.read_csv('/content/ME2N_2023.csv', sep=None, engine='python')
df_me2n_2024 = pd.read_csv('/content/ME2N_2024.csv', sep=None, engine='python')
df_me2n_2025 = pd.read_csv('/content/ME2N_2025.csv', sep=None, engine='python')
# sep=None will automatically detect the delimiter
# engine='python' is slower but more robust for complex cases
```

```
df_zmb51_2022_2025 = pd.read_csv('/content/zmb51_2022_2025_temiz.csv', sep=None, engine='python')
```

```
#ilk sutuna yıl yazdırma
df_me2n_2022["ГОД"] = "2022"
cols = ["ГОД"] + [col for col in df_me2n_2022.columns if col != "ГОД"]
df_me2n_2022 = df_me2n_2022[cols]
#
df_me2n_2023["ГОД"] = "2023"
cols = ["ГОД"] + [col for col in df_me2n_2023.columns if col != "ГОД"]
df_me2n_2023 = df_me2n_2023[cols]
#
df_me2n_2024["ГОД"] = "2024"
cols = ["ГОД"] + [col for col in df_me2n_2024.columns if col != "ГОД"]
df_me2n_2024 = df_me2n_2024[cols]
#
df_me2n_2025["ГОД"] = "2025"
cols = ["ГОД"] + [col for col in df_me2n_2025.columns if col != "ГОД"]
df_me2n_2025 = df_me2n_2025[cols]
```

```
df_listesi = [df_me2n_2022, df_me2n_2023, df_me2n_2024,df_me2n_2025]
```

```
birlesik_df = pd.concat(df_listesi, ignore_index=True)
```

```
birlesik_df.head()
```

```
len(birlesik_df)
```

```
birlesik_df["﻿Документ закупки"] = birlesik_df["﻿Документ закупки"].apply(temizle_ve_cevir)
birlesik_df["Позиция"] = birlesik_df["Позиция"].apply(temizle_ve_cevir)
birlesik_df["Закуп. организация"] = birlesik_df["Закуп. организация"].apply(temizle_ve_cevir)
birlesik_df["Заявка"] = birlesik_df["Заявка"].apply(temizle_ve_cevir)
```

```
birlesik_df = birlesik_df[birlesik_df["Инвентарный номер"].str.contains("AKU", na=False)]

```

```
len(birlesik_df)
```

```
birlesik_df.columns
```

```
birlesik_df.insert(0, 'zp-zppoz', birlesik_df['﻿Документ закупки'].astype(str) + birlesik_df['Позиция'].astype(str))
```

```
birlesik_df_zppozlar_unique = birlesik_df['zp-zppoz'].dropna().unique().tolist()
```

```
df_zmb51_2022_2025["Заказ на поставку"] = df_zmb51_2022_2025["Заказ на поставку"].apply(temizle_ve_cevir)
df_zmb51_2022_2025["Позиция"] = df_zmb51_2022_2025["Позиция"].apply(temizle_ve_cevir)
df_zmb51_2022_2025["Внутренний заказ"] = df_zmb51_2022_2025["Внутренний заказ"].apply(temizle_ve_cevir)
df_zmb51_2022_2025["Внутренний заказ"] = df_zmb51_2022_2025["Внутренний заказ"].apply(temizle_ve_cevir)
```

```
df_zmb51_2022_2025.insert(0, 'zp-zppoz', df_zmb51_2022_2025['Заказ на поставку'].astype(str) + df_zmb51_2022_2025['Позиция'].astype(str))
```

```
df_zmb51_2022_2025.columns
```

```
df_zmb51_2022_2025 = df_zmb51_2022_2025[
    df_zmb51_2022_2025['zp-zppoz'].isin(birlesik_df_zppozlar_unique)
].copy()
```
