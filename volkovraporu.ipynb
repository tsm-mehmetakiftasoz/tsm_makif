```
def temizle_ve_cevir(x):
    try:
        return str(int(float(x)))
    except:
        return str(x)
```

```
dfbirlesikmb52 = dfbirlesikmb52.rename(columns={"Инвентарный номер": "Инвентарный номер_old"})
```

```
# Önce mevcut Инвентарный номер kolonunu yeniden adlandır
dfbirlesikmb52 = dfbirlesikmb52.rename(columns={"Инвентарный номер": "Инвентарный номер_old"})

# dfbirlesikmb52 içindeki join anahtarı zaten var: Материал_Заказ_zmb52
# me2n tablosunda karşılık gelen anahtar: Материал_Заказ_me2n
# map ile yeni sütun oluştur
map_dict = dict(zip(me2n["Материал_Заказ_me2n"], me2n["Инвентарный номер"]))
map_dict2 = dict(zip(me2n["Материал_Заказ_me2n"], me2n["Позиция"]))

dfbirlesikmb52["inventory_number_2"] = dfbirlesikmb52["Материал_Заказ_zmb52"].map(map_dict)
dfbirlesikmb52["зп_Позиция"] = dfbirlesikmb52["Материал_Заказ_zmb52"].map(map_dict2)
# Sonuç kontrolü
dfbirlesikmb52[["Материал_Заказ_zmb52", "inventory_number_2","Заказ на поставку","зп_Позиция"]].head()
```

```
# Eğer inventory_number_2 boşsa ve Инвентарный номер_old doluysa, değeri kopyala
dfbirlesikmb52["inventory_number_2"] = dfbirlesikmb52["inventory_number_2"].fillna(
    dfbirlesikmb52["Инвентарный номер_old"]
)
```
