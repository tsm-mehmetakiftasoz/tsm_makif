```
# 1. Gerekli k√ºt√ºphaneleri y√ºkleyin ve i√ße aktarƒ±n
!pip install -q google-genai
import google.generativeai as genai
import ipywidgets as widgets
from IPython.display import display, clear_output

# 2. SABƒ∞TLERƒ∞N TANIMLANMASI (API Anahtarƒ± dahil)
# !!! Dƒ∞KKAT: API ANAHTARI KODA G√ñM√úLM√ú≈ûT√úR - G√úVENLƒ∞K Rƒ∞SKƒ∞ !!!
GOOGLE_API_KEY = "AIzaSyCYZP7v4C46Y-3TV-IYCO65JWxVMphUAPs"
MODEL_NAME = 'gemini-2.5-flash'
SIFIRLAMA_SIFRESƒ∞ = "muziplik_tamamlandƒ±"
SOZLESMELI_MESAJ = "Sohbeti Ba≈ülat" # ƒ∞lk butonda yazacak

# API Yapƒ±landƒ±rmasƒ±
try:
    genai.configure(api_key=GOOGLE_API_KEY)
except Exception as e:
    print(f"‚ùå Yapƒ±landƒ±rma Hatasƒ±: {e}")

# Global deƒüi≈ükenler
chat = None
output_area = widgets.Output() # Sohbet ge√ßmi≈üini tutmak i√ßin alan

def sohbeti_baslat_arayuz():
    """
    Sohbet oturumunu ba≈ülatƒ±r ve aray√ºz √∂ƒüelerini olu≈üturur.
    """
    global chat, output_area

    # Sohbeti temizle ve yeniden ba≈ülat
    clear_output(wait=True)

    try:
        model = genai.GenerativeModel(MODEL_NAME)
        chat = model.start_chat(history=[])

        with output_area:
            print("-" * 50)
            print(f"‚úÖ Gemini Sohbet Ba≈üarƒ±lƒ±! (Model: {MODEL_NAME})")
            print("üí° Mesajƒ±nƒ±zƒ± a≈üaƒüƒ±daki kutuya yazƒ±p 'G√∂nder'e tƒ±klayƒ±n.")
            print(f"√áƒ±kmak i√ßin: '{SIFIRLAMA_SIFRESƒ∞}' yazƒ±n.")
            print("-" * 50)

    except Exception as e:
        with output_area:
            print(f"‚ùå Ba≈ülatma Hatasƒ±: {e}")
        return

    # Aray√ºz √∂ƒüeleri
    metin_kutusu = widgets.Text(
        description='Mesajƒ±nƒ±z:',
        placeholder='Buraya yazƒ±n...',
        style={'description_width': 'initial'}
    )
    gonder_butonu = widgets.Button(description="G√∂nder")

    # D√ºƒüme tƒ±klandƒ±ƒüƒ±nda √ßalƒ±≈üacak fonksiyon
    def gonder_tusa_basildi(b):
        global chat
        kullanici_mesaji = metin_kutusu.value.strip()

        if not kullanici_mesaji:
            return

        # Output alanƒ±nƒ± kullanarak sohbeti g√ºncelle
        with output_area:
            # Sƒ±fƒ±rlama ≈üifresi kontrol√º
            if kullanici_mesaji == SIFIRLAMA_SIFRESƒ∞:
                print("\n‚úÖ ≈ûifre doƒüru. Sohbet oturumu sonlandƒ±rƒ±ldƒ±.")
                display(widgets.HTML('<hr>')) # Ayƒ±rƒ±cƒ± √ßizgi
                display(widgets.Label("Yeniden ba≈ülatmak i√ßin h√ºcreyi tekrar √ßalƒ±≈ütƒ±rƒ±n."))

                # Aray√ºz√º temizle ve durdur
                clear_output(wait=True)
                return

            print(f"\nSiz: {kullanici_mesaji}")

            try:
                # Mesajƒ± Gemini'ye g√∂nder ve yanƒ±t al
                response = chat.send_message(kullanici_mesaji)
                print(f"Gemini: {response.text}")
            except Exception as e:
                print(f"‚ùå Hata: Mesaj g√∂nderilemedi - {e}")

        # Metin kutusunu temizle
        metin_kutusu.value = ''

    # D√ºƒüme olayƒ±nƒ± baƒülama
    gonder_butonu.on_click(gonder_tusa_basildi)

    # Aray√ºz√º g√∂ster
    display(output_area, metin_kutusu, gonder_butonu)

# Fonksiyonu √ßaƒüƒ±rarak aray√ºz√º ba≈ülat
sohbeti_baslat_arayuz()
```
