```
def duzenle_sayi_sutunu(df, sutun_adi):
    """
    Sayısal değerleri metin olarak tutan ve binlik/ondalık ayracı içeren sütunu dönüştürür.
    - Nokta (.) binlik ayracı olarak kaldırılır
    - Virgül (,) ondalık ayracı olarak noktaya çevrilir
    - Float olarak dönüştürülür
    - Eğer tam sayı ise integer olarak, değilse float olarak kalır
    - Sonuç string olarak döndürülür (görsel temizlik için)

    Parametreler:
    df : pd.DataFrame
    sutun_adi : str

    Dönüş:
    df (işlenmiş hali)
    """
    df[sutun_adi] = (
        df[sutun_adi]
        .astype(str)
        .str.replace(" ", "", regex=False)
        .str.replace(".", "", regex=False)
        .str.replace(",", ".", regex=False)
        .astype(float)
        .apply(lambda x: str(int(x)) if x.is_integer() else str(x))
    )
    return df
```

```
def get_filtered_tables(df_calisilan, df_me5a_2022_2025, filtre_listesi):
    filtre_df = pd.DataFrame(filtre_listesi)

    # df_calisilan filtresi
    df_calisilan_filtre = df_calisilan.merge(
        filtre_df, on=["Заявка", "Позиция заявки"], how="inner"
    )[[
        "Заявка", "Позиция заявки", "Документ_закупки",
        "Зп_Поз", "Количество ЗП", "Итого цена нетто"
    ]]

    # df_me5a_2022_2025 filtresi
    df_me5a_filtre = df_me5a_2022_2025.merge(
        filtre_df, on=["Заявка", "Позиция заявки"], how="inner"
    )[[
        "Заявка", "Позиция заявки", "RU Наименование",
        "ЗатребованКолич","Заказанное к-во"
    ]]
    return df_calisilan_filtre, df_me5a_filtre
```
